```python
def load_metadata(root_path):
    """Load metadata from JSON files into a dict of dicts.

    Args:
        root_path (str): The root directory containing model subdirectories.

    Returns:
        dict: A dictionary with model IDs as keys and metadata dictionaries as values.
    """
    metadata_dict = {}
    for model_dir in os.listdir(root_path):
        model_path = os.path.join(root_path, model_dir)
        if os.path.isdir(model_path):
            metadata_file = os.path.join(model_path, f"{model_dir}_final_metadata.json")
            if os.path.isfile(metadata_file):
                with open(metadata_file, 'r') as file:
                    metadata_dict[model_dir] = json.load(file)
    return metadata_dict
```

```python
metadata = load_metadata(main_dir)
```

```python
import matplotlib.pyplot as plt

```

```python
# Extract width and height for each model
widths = [info["overall_bounding_box"]["width"] for info in metadata.values()]
heights = [info["overall_bounding_box"]["height"] for info in metadata.values()]

# Create scatter plot
plt.scatter(widths, heights)

# Label the axes
plt.xlabel('Width')
plt.ylabel('Height')

# Optional: Label each point with its model ID
for model_id, (width, height) in zip(metadata.keys(), zip(widths, heights)):
    plt.annotate(model_id, (width, height))

# Show the plot
plt.show()
```
