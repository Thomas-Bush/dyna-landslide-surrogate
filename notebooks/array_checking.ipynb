```python
import numpy as np
import matplotlib.pyplot as plt
import os

def plot_arrays_from_files(file_path1, file_path2):
    """
    Plots arrays from two files side by side.

    Parameters:
    - file_path1: Path to the first .npy file.
    - file_path2: Path to the second .npy file.
    """

    # Load the arrays
    array1 = np.load(file_path1)
    array2 = np.load(file_path2)

    # Check if the arrays have the expected shape
    if array1.shape == (256, 256) and array2.shape == (256, 256):
        # Create a figure with 2 subplots side by side
        fig, axs = plt.subplots(1, 2, figsize=(10, 5))

        # Plot the first array
        axs[0].imshow(array1, cmap='viridis')
        axs[0].set_title(f'Velocity')
        axs[0].axis('off')  # Hide the axis

        # Plot the second array
        axs[1].imshow(array2, cmap='viridis')
        axs[1].set_title(f'Thickness')
        axs[1].axis('off')  # Hide the axis

        plt.show()
    else:
        print("One or both arrays do not have the expected shape (256x256).")

```
