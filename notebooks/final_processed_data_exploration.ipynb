```python
import numpy as np
import glob
import os
from matplotlib import pyplot as plt

# Initialize lists to store statistics
non_zero_percentages = []
bounding_boxes = []

# Define the data directory path
data_dir = 'C:\\Users\\thomas.bush\\repos\\dyna-landslide-surrogate\\data'

# Loop through each model directory
for model_dir in glob.glob(os.path.join(data_dir, '*')):
    if os.path.isdir(model_dir):
        # Define the pattern to match the .npy files
        file_pattern = os.path.join(model_dir, '04_FinalProcessedData_256', 'velocity', '*_velocity_*.npy')
        
        # Loop through each .npy file in the current model's directory
        for file_path in glob.glob(file_pattern):
            # Load the array from the .npy file
            arr = np.load(file_path)
            
            # Calculate the percentage of non-zero cells
            non_zero_count = np.count_nonzero(arr)
            total_cells = arr.size
            non_zero_percentage = (non_zero_count / total_cells) * 100
            non_zero_percentages.append(non_zero_percentage)
            
            # Find the bounding box of non-zero cells if there are any
            non_zero_indices = np.argwhere(arr > 0)
            if non_zero_indices.size > 0:
                min_row, min_col = np.min(non_zero_indices, axis=0)
                max_row, max_col = np.max(non_zero_indices, axis=0)
                bounding_box = ((max_col - min_col + 1), (max_row - min_row + 1))
                bounding_boxes.append(bounding_box)
            else:
                # Handle the case where there are no non-zero cells
                bounding_boxes.append((0, 0))




# Plotting the histograms
# Histogram of non-zero cell percentages
plt.figure(figsize=(10, 4))
plt.hist(non_zero_percentages, bins=50, color='blue', alpha=0.7)
plt.title('Histogram of Non-Zero Cell Percentages')
plt.xlabel('Percentage of Non-Zero Cells')
plt.ylabel('Frequency')
plt.show()

# Histogram of bounding box widths
plt.figure(figsize=(10, 4))
box_widths = [box[0] for box in bounding_boxes]
plt.hist(box_widths, bins=50, color='green', alpha=0.7)
plt.title('Histogram of Bounding Box Widths')
plt.xlabel('Width in Pixels')
plt.ylabel('Frequency')
plt.show()

# Histogram of bounding box heights
plt.figure(figsize=(10, 4))
box_heights = [box[1] for box in bounding_boxes]
plt.hist(box_heights, bins=50, color='red', alpha=0.7)
plt.title('Histogram of Bounding Box Heights')
plt.xlabel('Height in Pixels')
plt.ylabel('Frequency')
plt.show()
```

```python
import numpy as np
import glob
import os
from concurrent.futures import ThreadPoolExecutor

# Function to load the array and check if it is empty
def is_array_empty(file_path):
    arr = np.load(file_path)
    return arr.size == 0 or np.count_nonzero(arr) == 0

# Define the data directory path
data_dir = 'C:\\Users\\thomas.bush\\repos\\dyna-landslide-surrogate\\data'

# Get a list of all model directories
model_dirs = [d for d in glob.glob(os.path.join(data_dir, '*')) if os.path.isdir(d)]

# Initialize counters
total_arrays_checked = 0
total_empty_arrays = 0

# Initialize a dictionary to hold the results
empty_arrays_info = {}

# Using ThreadPoolExecutor to handle IO-bound tasks
with ThreadPoolExecutor() as executor:
    for model_dir in model_dirs:
        # Define the pattern to match the .npy files
        file_pattern = os.path.join(model_dir, '04_FinalProcessedData_128', 'velocity', '*_velocity_*.npy')
        
        # Get the file paths
        file_paths = glob.glob(file_pattern)
        
        # Map the is_array_empty function to all files found by the glob pattern
        results = list(executor.map(is_array_empty, file_paths))
        
        # Update the counters based on the results
        total_arrays_checked += len(file_paths)
        total_empty_arrays += results.count(True)
        
        # Update the dictionary with the results
        empty_arrays_info[model_dir] = [file_path for file_path, is_empty in zip(file_paths, results) if is_empty]

# Print out the total number of arrays checked and total number of empty arrays
print(f"Total number of arrays checked: {total_arrays_checked}")
print(f"Total number of empty arrays: {total_empty_arrays}")

# Now you can print out or process the empty arrays info
for model_dir, empty_files in empty_arrays_info.items():
    print(f"Model Directory: {model_dir}")
    for file_path in empty_files:
        print(f"Empty array found in file: {file_path}")
```

```python
# Return keys that have non-empty values
non_empty_keys = [key for key, value in empty_arrays_info.items() if value]

# Print the non-empty keys
print("Model directories with empty arrays:")
print(non_empty_keys)
```
